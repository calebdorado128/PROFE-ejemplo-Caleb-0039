```python
# clase Automovil, Atributos: marca, placa, precio
# Funciones: leer_datos() y mostrar_datos()
# crear 3 objetos ---> carro, carro2 y carro3
# Leer datos

class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leer_datos(self):
        self.marca = input("Ingrese la marca: ")
        self.placa = input("Ingrese la placa: ")
        self.precio = input("Ingrese el precio: ")

    def mostrar_datos(self):
        print("Marca:", self.marca)
        print("Placa:", self.placa)
        print("Precio:", self.precio)
        return self.marca, self.placa, self.precio
# fin de la clase Automovil

print("============================")
print("------ Carro Numero 1 ------")
carro1 = Automovil("", "", "") # creando objeto carro 1
carro1.leer_datos()
carro1.mostrar_datos()
print("============================")
print("------ Carro Numero 2 ------")
carro2 = Automovil("", "", "") # creando objeto carro 2
carro2.leer_datos()
carro2.mostrar_datos()
print("============================")
print("------ Carro Numero 3 ------")
carro3 = Automovil("", "", "") # creando carro objeto 3
carro3.leer_datos()
carro3.mostrar_datos()
print("===========================")
print("Software realizado por Caleb Dorado, NL = 14")
```
